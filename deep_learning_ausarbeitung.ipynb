Mit Hilfe der Funktion plot_transformed_image() können wir den Unterschied zwischen den transformierten und den normalisierten Bildern sehen.

```python
def plot_transformed_bild(image_paths, transform, n=3):

    random_image_paths = random.sample(image_paths, k=n)
    for image_path in random_image_paths:
        with Image.open(image_path) as f:
            fig, ax = plt.subplots(1, 2)
            ax[0].imshow(f) 
            ax[0].set_title(f"Original \nSize: {f.size}")
            ax[0].axis("off")

            transformed_image = transform(f).permute(1, 2, 0) 
            ax[1].imshow(transformed_image) 
            ax[1].set_title(f"Transformed \nSize: {transformed_image.shape}")
            ax[1].axis("off")

            fig.suptitle(f"Class: {image_path.parent.stem}", fontsize=16)

plot_transformed_bild(image_path_list, 
                        transform=train_transforms, 
                        n=3)
```

## 6. Implementierung der Lösung mit PyTorch

In diesem Abschnitt konzentrieren wir uns auf die praktische Umsetzung von zwei verschiedenen Convolutional Neural Networks (CNN) zur Lösung der Bildklassifizierung: SimpleCNN und TinyVGG. Diese Modelle wurden ausgewählt, um einen Einblick in die Bandbreite der Möglichkeiten zu geben, die moderne Deep Learning Frameworks wie PyTorch bieten, um leistungsfähige Bildklassifizierungssysteme zu entwickeln und zu trainieren.

### 6.1 SimpleCNN Modell

Das SimpleCNN Modell ist ein einfaches, gefaltetes neuronales Netz, das aus einigen gefalteten Schichten besteht, auf die vollständig verbundene Schichten folgen.

### 

```python
import torch.nn as nn
import torch.nn.functional as F

# Definition des CNN-Modells
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # Eingabebild: 3 Kanäle, Ausgabe: 6 Kanäle, 5x5 quadratisches Faltungsfenster
        self.conv1 = nn.Conv2d(3, 6, 5)
        # Pooling über ein 2x2 Fenster
        self.pool = nn.MaxPool2d(2, 2)
        # Zweite Faltungsschicht: 6 Eingabekanäle, 16 Ausgabekanäle, 5x5 quadratisches Faltungsfenster
        self.conv2 = nn.Conv2d(6, 16, 5)
        # Drei Fully-Connected Layer
        # Ein Bild wird durch das Netz auf eine Größe von 5x5x16 reduziert (nach zwei Conv- und zwei Pooling-Layern)
        self.fc1 = nn.Linear(16 * 13 * 13, 120)
        self.fc2 = nn.Linear(120, 84)
        # Drei Ausgabekanäle für die drei Klassen
        self.fc3 = nn.Linear(84, 3)

    def forward(self, x):
        # Anwenden von Convolution, Activation (ReLU) und Pooling
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        # Umwandlung der Daten für Fully-Connected Layer
        x = x.view(-1, 16 * 13 * 13)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


torch.manual_seed(42)
simple_cnn = SimpleCNN()

simple_cnn
```

Das Netzwerk beginnt mit einer Convolutional Layer, die ein Eingangsbild mit 3 Kanälen aufnimmt und mit einem 5x5-Filter 6 Ausgangskanäle erzeugt, gefolgt von einer Max Pooling Schicht mit einem 2x2-Fenster zur Reduzierung der räumlichen Dimensionen. Eine zweite Convolutional Layer erhöht die Tiefe auf 16 Kanäle mit einem weiteren 5x5-Filter und wird wiederum von einer Max-Pooling-Schicht gefolgt. Nach der Verarbeitung durch die Faltungsschicht und die Pooling-Schicht wird das Bild auf eine Größe von 5x5x16 reduziert. Es folgen drei voll verknüpfte (dichte) Schichten, die die Merkmale auf 120, dann auf 84 Einheiten reduzieren, und schließlich eine Ausgabeschicht mit 3 Einheiten zur Klassifizierung in drei Klassen. Die Aktivierungsfunktion ReLU wird nach den Faltungsschichten und den ersten beiden dichten Schichten verwendet, um Nichtlinearitäten einzuführen und es dem Netz zu ermöglichen, komplexe Muster in den Daten zu lernen.

### 6.2 TinyVGG Modell

Das TinyVGG-Modell ist eine vereinfachte Version des VGG-Netzwerks, die für schnelle Experimente geeignet ist. Es behält die Grundlegende Idee der Verwendung mehrerer Faltungsschichten bei, reduziert jedoch die Tiefe und Komplexität des Netzwerks.

```python
class TinyVGG(nn.Module):
    """
    Modellarchitektur, von der TinyVGG kopiert wird: [3]
    """
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape, 
                      out_channels=hidden_units, 
                      kernel_size=3, # how big is the square that's going over the image?
                      stride=1, # default
                      padding=1), # options = "valid" (no padding) or "same" (output has same shape as input) or int for specific number 
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units, 
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2,
                         stride=2) # default stride value is same as kernel_size
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            # Where did this in_features shape come from? 
            # It's because each layer of our network compresses and changes the shape of our inputs data.
            nn.Linear(in_features=hidden_units*16*16,
                      out_features=output_shape)
        )
    
    def forward(self, x: torch.Tensor):
        x = self.conv_block_1(x)
        # print(x.shape)
        x = self.conv_block_2(x)
        # print(x.shape)
        x = self.classifier(x)
        # print(x.shape)
        return x

torch.manual_seed(42)
tiny_vgg = TinyVGG(input_shape=3, # number of color channels (3 for RGB) 
                  hidden_units=10, 
                  output_shape=len(train_data.classes))
tiny_vgg
```

Die Klasse nimmt als Eingabeparameter die Dimensionen der Eingabebilder (input_shape), die Anzahl der verborgenen Einheiten (hidden_units) und die Anzahl der Ausgabeklassen (output_shape). Das Netzwerk besteht aus zwei konvolutionellen Blöcken (conv_block_1 und conv_block_2), die jeweils zwei Convolutional Layers mit ReLU-Aktivierungsfunktionen und nachfolgendem MaxPooling enthalten, um die räumlichen Dimensionen der Merkmalskarten zu reduzieren. Nach Durchlaufen dieser konvolutionellen Blöcke werden die Merkmalsdaten abgeflacht und durch einen Fully-Connected Layer (classifier) geführt, der die finale Klassifikation vornimmt. Die Architektur zielt darauf ab, effektive Merkmalsextraktion für Bildklassifizierungsaufgaben zu bieten, indem sie wichtige visuelle Merkmale durch sukzessive Filterung und räumliche Reduktion erfasst. 

## 7. Modell trainieren und evaluieren

Um die beide Modelle (SimpleCNN und TinyVGG)  nacheinander zu trainieren und dann sowohl die Genauigkeit als auch den F1-Score für jedes Modell auf den Testdaten berechnet, können wir die Funktion train_and_evaluate_model() verwenden.  Diese Funktion nimmt als Argumente die Trainings- und Testdaten-Loader sowie die Modellkonfigurationen entgegen und führt das Training und die Evaluierung für jedes Modell durch. Nach dem Training jedes Modells werden die Leistungsmetriken ausgegeben.

```python
from sklearn.metrics import f1_score
from timeit import default_timer as timer
import torch
import torch.optim as optim
import torch.nn.functional as F
from torch import nn

def train_and_evaluate_model(model, train_dataloader, test_dataloader, epochs=10):
    # Modell, Optimierer und Verlustfunktion initialisieren
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    # Trainingsschleife
    model.train()
    start_time = timer()
    for epoch in range(epochs):
        for images, labels in train_dataloader:
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        print(f'Epoch {epoch+1}, Loss: {loss.item()}')
    end_time = timer()

    # Evaluierungsschleife
    model.eval()
    total = 0
    correct = 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in test_dataloader:
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    accuracy = 100 * correct / total
    f1score = f1_score(all_labels, all_preds, average='weighted')

    print(f"Total training time for {model.__class__.__name__}: {end_time - start_time:.3f} seconds")
    print(f'Accuracy of {model.__class__.__name__} on the test images: {accuracy}%')
    print(f'F1-Score of {model.__class__.__name__}: {f1score}\n')
    return accuracy, f1score
```

```python
torch.manual_seed(42) 
torch.cuda.manual_seed(42)
simple_cnn_model = simple_cnn
tiny_vgg_model = tiny_vgg  

# Trainieren und Evaluieren der Modelle
print("Training und Evaluierung von SimpleCNN:")
simple_cnn_model = train_and_evaluate_model(simple_cnn_model, train_dataloader, test_dataloader)

print("Training und Evaluierung von TinyVGG:")
tiny_vgg_model = train_and_evaluate_model(tiny_vgg_model, train_dataloader, test_dataloader)

simple_cnn_model, tiny_vgg_model
```

```python
import random
import torchvision
# Funktion zum Anzeigen eines Bildes
def imshow(img):
    img = img / 2 + 0.5  # Unnormalisieren
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# Zufälliges Bild aus dem Testdatensatz holen
images, labels = next(iter(test_dataloader))

# Wähle ein zufälliges Bild
index = random.randrange(len(images))
img = images[index]

# Bild anzeigen
imshow(torchvision.utils.make_grid(img))

# Vorhersage des Model SimpleCNN
outputs_cnn = simple_cnn(img.unsqueeze(0)) 
_, predicted_cnn = torch.max(outputs_cnn, 1)


# Vorhersage des Model SimpleCNN
outputs_tinyVGG = tiny_vgg(img.unsqueeze(0)) 
_, predicted_tinyVGG = torch.max(outputs_tinyVGG, 1)

print(f'Vorhergesagte Klasse des SimpleCNN: {test_data.classes[predicted_cnn[0]]}')
print(f'Vorhergesagte Klasse des TinyVGG: {test_data.classes[predicted_tinyVGG[0]]}')
```
